```python
# Import necessary libraries
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Load the saved model (.keras format)
model = load_model(r'C:\Users\rohit\lunglens2\vgg16_pneumonia_detection.keras')

# Function to load and preprocess a single image
def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))  # Load image
    img_array = image.img_to_array(img)  # Convert image to array
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array /= 255.0  # Rescale pixel values
    return img_array

# Function to make a prediction on a single image
def predict_image(img_path):
    img_array = load_and_preprocess_image(img_path)  # Preprocess the image
    prediction = model.predict(img_array)  # Make a prediction
    class_label = 'Pneumonia' if prediction[0][0] > 0.5 else 'Normal'  # Classify based on prediction
    confidence = prediction[0][0] if prediction[0][0] > 0.5 else 1 - prediction[0][0]  # Calculate confidence
    return class_label, confidence

# Example usage: select an image for prediction
img_path = 'C:\\Users\\rohit\\lunglens2\\chest_xray\\test\\PNEUMONIA\\person20_virus_51.jpeg'  # Change this to the path of the image you want to test
class_label, confidence = predict_image(img_path)

# Display the result
print(f'Predicted Class: {class_label}')
print(f'Confidence: {confidence:.2f}')

# Optionally, display the image
img = image.load_img(img_path, target_size=(224, 224))
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()
```
