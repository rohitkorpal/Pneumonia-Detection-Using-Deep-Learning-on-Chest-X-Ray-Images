```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import gradio as gr

# Load your Keras model
model_path = '/mnt/data/vgg16_pneumonia_detection.keras'  # Path to the uploaded model
model = load_model(model_path)

# Function to preprocess the image
def preprocess_image(img):
    img = img.resize((224, 224))  # Resize to the input size required by the model
    img_array = np.array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0  # Normalize the image
    return img_array

# Prediction function to be used with Gradio
def predict(img):
    img_array = preprocess_image(img)
    prediction = model.predict(img_array)
    
    # Assuming binary classification (0 for Normal, 1 for Pneumonia)
    if prediction[0] > 0.5:
        result = "Pneumonia Detected"
    else:
        result = "No Pneumonia"
    
    return result

# Gradio interface
interface = gr.Interface(
    fn=predict,  # Function to run
    inputs="image",  # Input type: image
    outputs="text",  # Output type: text
    title="Pneumonia Detection with VGG16",  # Title of the interface
    description="Upload an X-ray image to detect Pneumonia."
)

# Launch the Gradio interface
interface.launch()
```
